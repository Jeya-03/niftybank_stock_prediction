LSTM

```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt

# Function to retrieve data for a given symbol with only open and close prices
def get_stock_data(symbol, start_date, end_date):
    try:
        data = yf.download(symbol, start=start_date, end=end_date)[['Open', 'Close']]
        return data
    except Exception as e:
        print(f"Failed to download data for {symbol}: {e}")
        return None

# Function to preprocess data and split into training and testing sets
def preprocess_data(data, train_percentage=0.8):
    if data is None or data.empty:
        return None, None, None

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)

    train_size = int(len(scaled_data) * train_percentage)
    if train_size == 0:
        return None, None, None

    train_data = scaled_data[:train_size]
    test_data = scaled_data[train_size:]

    return train_data, test_data, scaler

# Function to create LSTM model with refined architecture
def create_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=128, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(Dense(units=2))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Function to train LSTM model with adjusted epochs and batch size for higher accuracy
def train_model(model, X_train, y_train):
    model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=2)

# Function to evaluate model performance
def evaluate_model(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    accuracy = r2_score(y_true, y_pred)
    return mse, accuracy

# Function to plot predicted vs actual prices for a stock
def plot_predictions(symbol, actual_prices, predicted_prices, dates, scaler):
    actual_prices = scaler.inverse_transform(actual_prices)
    predicted_prices = scaler.inverse_transform(predicted_prices)

    plt.figure(figsize=(16, 8))
    plt.plot(dates, actual_prices[:, 1], label='Actual Close Prices')
    plt.plot(dates, predicted_prices[:, 1], label='Predicted Close Prices')
    plt.title(f"Actual vs Predicted Close Prices for {symbol}")
    plt.xlabel("Date")
    plt.ylabel("Close Price (INR)")
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

# Main function with further refined settings for improved accuracy
def main():
    symbols =  ['ICICIBANK.NS', 'AXISBANK.NS', 'SBIN.NS', 'INDUSINDBK.NS', 'FEDERALBNK.NS', 'IDFCFIRSTB.NS', 'PNB.NS', 'BANKBARODA.NS']
    end_date = input("Enter end date (YYYY-MM-DD): ")
    end_date_dt = pd.to_datetime(end_date)
    start_date_dt = pd.to_datetime('2000-01-01')

    results = []

    for symbol in symbols:
        print(f"Processing {symbol}...")
        data = get_stock_data(symbol, start_date=start_date_dt, end_date=end_date)
        if data is None:
            print(f"No data available for {symbol}. Skipping...")
            continue

        train_data, test_data, scaler = preprocess_data(data)
        if train_data is None or test_data is None or scaler is None:
            print(f"Insufficient data for {symbol}. Skipping...")
            continue

        X_train, y_train = train_data[:-1], train_data[1:]
        X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))  # Reshape for LSTM input
        X_test, y_test = test_data[:-1], test_data[1:]
        X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))  # Reshape for LSTM input

        model = create_lstm_model(input_shape=(1, 2))
        train_model(model, X_train, y_train)

        predictions = model.predict(X_test)

        mse, accuracy = evaluate_model(y_test, predictions)

        dates = data.index[-len(predictions):]  # Dates for the test data

        plot_predictions(symbol, test_data[1:], predictions, dates, scaler)

        results.append({
            'Symbol': symbol,
            'Predicted Close Price': scaler.inverse_transform(predictions)[-1][1],
            'Accuracy (%)': accuracy * 100,
            'MSE': mse,
            'R-squared': accuracy
        })
    if results:
        results_df = pd.DataFrame(results)
        print(results_df)

if __name__ == "__main__":
    main()
```
